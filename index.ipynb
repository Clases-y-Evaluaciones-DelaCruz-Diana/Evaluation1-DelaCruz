```python
#PRIMERO: Se llaman a los excels que se subieron al folder del Jupiter y se les asigna un Data Frame
import pandas as pd

gastopublicLink='https://github.com/Clases-y-Evaluaciones-DelaCruz-Diana/Evaluation1-DelaCruz/raw/main/excelsdevariables/GastopublicoEducacion.csv'
onuLink='https://github.com/Clases-y-Evaluaciones-DelaCruz-Diana/Evaluation1-DelaCruz/raw/main/excelsdevariables/ONU.csv'
reservaspetLink='https://github.com/Clases-y-Evaluaciones-DelaCruz-Diana/Evaluation1-DelaCruz/raw/main/excelsdevariables/Reservaspet.csv'

gastopublic=pd.read_csv(gastopublicLink)
onu=pd.read_csv(onuLink)
reservaspet=pd.read_csv(reservaspetLink)
```

```python
#SEGUNDO: Se revisa como esta compuesta su matriz por filas y columnas
gastopublic.shape,onu.shape,reservaspet.shape
```

```python
#TERCERO: Se revisa el nombre de las cabeceras de cada columna de los Data Frames
gastopublic.columns,onu.columns,reservaspet.columns
```

```python
#CUARTO: Se unen los tres Data Frames anteriores para formar un solo DF que contenga las columnas de las variables elegidas 
merge1=gastopublic.merge(onu,how='inner',left_on='Countries',right_on='Countries').merge(reservaspet,how='inner',left_on='Countries',right_on='Countries')
merge1
```

```python
#QUINTO: Se revisa el numero de filas y columnas como resultado de la union
merge1.shape
```

```python
#SEXTO: Se reviso previamente el .csv que entrego el profesor, se observo una diferencia en la cabecera de la columna paises
```

```python
#SETIMO: La diferencia consiste en que la cabecera del primer merge era plural, mientras que la otra era singular
# Por lo tanto, se renombra la columna del primer merge a fin de que coincida con la columna del .csv del profesor
merge1=merge1.rename(columns={"Countries": "Country"})
merge1
```

```python
#OCTAVO: Tambien se observa que en la DF del profesor, la columna Country posee elementos en mayuscula
# Por lo tanto, el merge1 se convierte a mayusculas para poder fusionarse con la DF del profesor
merge1['Country']= merge1['Country'].str.upper()
merge1
```

```python
merge1.Country=merge1.Country.str.strip()
merge1
```

```python
#DECIMO: Se incorpora el .csv del profesor y se le asigna un DF
import pandas as pd

countryisosLink='https://github.com/Clases-y-Evaluaciones-DelaCruz-Diana/Evaluation1-DelaCruz/raw/main/excelsdevariables/country_isos.csv'

countryisos=pd.read_csv(countryisosLink)
```
